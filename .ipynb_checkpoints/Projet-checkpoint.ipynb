## Variables du dataset churn_dataset.csv

1. RowNumber : Numéro de ligne (inutile pour l’analyse, à supprimer).
2. CustomerId : Identifiant unique du client (inutile pour l’analyse, à supprimer).
3. Surname : Nom du client (inutile pour l’analyse, à supprimer).
4. CreditScore : Score bancaire attribué au client (numérique).
5. Geography : Pays où réside le client (catégorique, à encoder).
6. Gender : Genre du client (catégorique, à encoder : Male/Female).
7. Age : Âge du client (numérique).
8. Tenure : Nombre d'années où le client est resté avec la banque (numérique).
9. Balance : Solde du compte bancaire du client (numérique).
10. NumOfProducts : Nombre de produits bancaires utilisés par le client (numérique).
11. HasCrCard : Indique si le client possède une carte de crédit (1 = Oui, 0 = Non).
12. IsActiveMember : Indique si le client est un membre actif (1 = Oui, 0 = Non).
13. EstimatedSalary : Salaire estimé du client (numérique).
14. Exited : Variable cible (1 = Client sur le départ, 0 = Client qui reste).

```python
# Classification des colonnes par type
continuous_vars = []  # Variables continues
categorical_vars = []  # Variables catégoriques
binary_vars = []  # Variables binaires

for column in df.columns:
    unique_values = df[column].nunique()  # Nombre de valeurs uniques
    if df[column].dtype in ['int64', 'float64']:  # Type numérique
        if unique_values == 2:  # Seulement 2 valeurs uniques
            binary_vars.append(column)
        else:  # Plus de 2 valeurs uniques
            continuous_vars.append(column)
    else:  # Type non numérique
        categorical_vars.append(column)

# Affichage des résultats
print("Variables continues :", continuous_vars)
print("Variables catégoriques :", categorical_vars)
print("Variables binaires :", binary_vars)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from xgboost import XGBClassifier
import pandas as pd

# Charger les données
df = pd.read_csv('churn_dataset.csv')

# Définir les colonnes et la cible
target = 'Exited'
cont_columns  = ['CreditScore', 'Age', 'Tenure', 'Balance', 'EstimatedSalary']
cat_columns = ['Geography', 'Gender']
binary_columns = ['HasCrCard', 'IsActiveMember', 'NumOfProducts']

# Définir X et y
X = df[cat_columns + cont_columns + binary_columns]  # Variables explicatives
y = df[target]  # Variable cible

# Diviser les données en train/test
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Prétraitement des données
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), cat_columns),
        ('binary', 'passthrough', binary_columns),
        ('num', StandardScaler(), cont_columns)
    ]
)

# Configurer SMOTE et le pipeline
smote = SMOTE(random_state=42)

model = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', smote),
    ('classifier', XGBClassifier(learning_rate=0.1, max_depth=3, n_estimators=200, random_state=42))
])

# Entraîner le modèle
model.fit(X_train, Y_train)

# Prédictions
y_pred = model.predict(X_test)

# Évaluer les performances
print(f"Accuracy: {accuracy_score(Y_test, y_pred)}")
print("\nRapport de classification :\n", classification_report(Y_test, y_pred))
```
